5 DAY SIMPLE MOVING AVERAGE

```python
import csv

p = 5  ### select the window for moving average
date_dict={}
path_to_dataset = '/content/drive/MyDrive/SLDC_Data/2023/01/'  ## import data from google drive

### select the dates for which you wish to make the moving average 
dates = ['01-01-2023.csv','02-01-2023.csv','03-01-2023.csv','04-01-2023.csv','05-01-2023.csv','06-01-2023.csv']
```

```python
###### read data from individual files
for i in range(len(dates)):
    date_dict[dates[i]] = []
    time = []
    for d in csv.DictReader(open(path_to_dataset+dates[i]), delimiter=','):
        time.append(d['time'])
        date_dict[dates[i]].append(float(d['value']))
n = len(date_dict[dates[len(dates)-1]])


## take moving average of for each day
forecast =[0]*n
for i in range(n):
    count =0
    for j in range(p):
      try:
        forecast[i] += date_dict[dates[j]][i] 
        count +=1
      except:
        pass 
    print(count, forecast[i])
    forecast[i]=forecast[i]/count
```

Print Prediced with Actual Load Values for each day

```python
for i in range(n):
    print(time[i]," : ",date_dict[dates[len(dates)-1]][i]," : ",forecast[i])
```

Draw Load Curve of Atual and Predicted Values

```python
plt.plot(date_dict[dates[len(dates)-1]])
plt.plot(forecast)
plt.legend(['load', 'forecast'])
plt.show()
```
