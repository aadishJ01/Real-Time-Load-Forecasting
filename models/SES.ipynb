```python
import csv

date_dict={}
path_to_dataset = '/content/drive/MyDrive/SLDC_Data/2023/01/'

dates = ['01-01-2023.csv','02-01-2023.csv','03-01-2023.csv','04-01-2023.csv','05-01-2023.csv','06-01-2023.csv','07-01-2023.csv','08-01-2023.csv','09-01-2023.csv','10-01-2023.csv','11-01-2023.csv','12-01-2023.csv','13-01-2023.csv','14-01-2023.csv','15-01-2023.csv','16-01-2023.csv','17-01-2023.csv','18-01-2023.csv','19-01-2023.csv','20-01-2023.csv','21-01-2023.csv','22-01-2023.csv','23-01-2023.csv','24-01-2023.csv','25-01-2023.csv','26-01-2023.csv','27-01-2023.csv','28-01-2023.csv','29-01-2023.csv','30-01-2023.csv']
for i in range(len(dates)):
    date_dict[dates[i]] = []
    time = []
    for d in csv.DictReader(open(path_to_dataset + dates[i]), delimiter=','):
        time.append(d['time'])
        date_dict[dates[i]].append(float(d['value']))

for i in range(len(dates)):
    length = len(date_dict[dates[i]])
    value = date_dict[dates[i]][length-1]
    for j in range(length,288):
        date_dict[dates[i]].append(value)

n = len(date_dict[dates[len(dates)-1]])
m = len(dates)
# print(n,m)
alpha  = [0]*n
alphamin = ['value']+[0.1]*n
forecast =[[0]*n for i in range(m)]
forecast[1] = [date_dict[dates[0]][j] for j in range(n)]
for j in range(n):
    mse = [0]*9
    for k in range(1,10):
        alpha[j] = k*0.1
        for i in range(2,m):
            # print(i,j,k)
            forecast[i][j] = (alpha[j]*date_dict[dates[i-1]][j]) + ((1 - alpha[j])*forecast[i-1][j])
        for i in range(1,m):
            mse[k-1] += (forecast[i][j] - date_dict[dates[i]][j])**2
    min = mse[0]
    for i in range(1,9):
        if mse[i]<min:
            min = mse[i]
            alphamin[j+1] = (i+1)*0.1
print(alphamin)
```
